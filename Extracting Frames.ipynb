```python
import os

# Define the root directory where your folders are located
root_directory = '/home/abi/Hard Disk/Final-Year-Project/Elephant-Dataset/Videos'

# Initialize an empty list to store video file paths
video_files = []

# Recursively search for video files in all subdirectories
for root, dirs, files in os.walk(root_directory):
    for file in files:
        # Check if the file has a video file extension (e.g., .mp4, .avi, .mov, etc.)
        if file.lower().endswith(('.mp4', '.webm')):
            # Append the full file path to the video_files list
            video_files.append(os.path.join(root, file))

# Print the list of video file paths
for video_file in video_files:
    print(video_file)

# total number of videos
print(len(video_files))
```

```python
import cv2
import os

# Function to extract frames from a video and save them as images
def extract_frames(video_path, output_dir, frame_rate=1):
    # Create the output directory if it doesn't exist
    global file_count
    os.makedirs(output_dir, exist_ok=True)

    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Get the frames per second (FPS) of the video
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Set the frame counter and file counter to 0
    frame_count = 0

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # Capture frames at the specified frame_rate
        if frame_count % (fps // frame_rate) == 0:
            # Save the frame as an image
            print(f'frame_{file_count:04d}.jpg')
            frame_filename = os.path.join(output_dir, f'frame_{file_count:04d}.jpg')
            cv2.imwrite(frame_filename, frame)
            file_count += 1

        frame_count += 1

    # Release the video capture object
    cap.release()

# Output directory for the extracted frames
output_directory = 'frames_output'
file_count = 0

# Process each video file
for video_file in video_files:
    video_path = os.path.join('videos', video_file)  
    # Adjust the path to your video files
    extract_frames(video_path, output_directory)

print("Frames extraction completed.")
```
